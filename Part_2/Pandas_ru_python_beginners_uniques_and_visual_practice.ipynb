```python
df_uniques = pd.melt(frame=df, value_vars=['Messages','Active days', 
                                                 'Flood ratio'])


df_uniques = pd.DataFrame(df_uniques.groupby(['variable', 
                                                    'value'])['value'].count()) \
    .sort_index(level=[0, 1]) \
    .rename(columns={'value': 'count'}) \
    .reset_index()
    
sns.factorplot(x='variable', y='count', hue='variable', 
               data=df_uniques, kind='bar', size=12);
```

```python
df = pd.read_csv("C:/Users/Battlestation/Downloads/chat.csv")
```

```python
corr = df.corr(method='pearson')

# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –º–∞—Å–∫—É –¥–ª—è –æ—Ç—Å–µ—á–µ–Ω–∏—è –≤–µ—Ä—Ö–Ω–µ–π –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –º–∞—Ç—Ä–∏—Ü—ã
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(12, 9))

sns.heatmap(corr, mask=mask, vmax=1, center=0, annot=True, fmt='.1f',
            square=True, linewidths=.5, cbar_kws={"shrink": .5});   
```
